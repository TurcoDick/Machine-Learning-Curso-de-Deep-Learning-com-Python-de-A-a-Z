```python
import pandas as pd
from sklearn.model_selection import train_test_split
from keras.models import Sequential
from keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
from keras.utils import np_utils
from sklearn.metrics import confusion_matrix
import numpy as np
```

```python
base = pd.read_csv('../datasets/iris.csv')
```

```python
previsores = base.iloc[:, 0:4].values
```

```python
classe = base.iloc[:, 4].values
```

```python
classe
```

## transformando os valores de classe de categóricos para númericos ##

```python
labelencoder = LabelEncoder()
```

```python
classe = labelencoder.fit_transform(classe)
```

```python
classe
```

   ## Ainda não esta pronto, as saídas terão três valores, pois são três neuronios de saída, então ## 

   #### iris setosa            1 0 0 ###
   #### iris virginica  0 1 0 ###
   #### iris versicolor 0 0 1 ###


```python
classe_dummy = np_utils.to_categorical(classe)
classe_dummy
```

## separando em treinamento e teste ##

```python
previsores_treinamento, previsores_teste, classe_treinamento, classe_teste = train_test_split(
    previsores, classe_dummy, test_size= 0.25)
```

## Criando a rede neural, o units é a quantidade de neuronios na câmada oculta (units),##
## e o cálculo é qtde de colunas + qtde de classe dividido por 2 ##

```python
rede = Sequential()
rede.add(Dense(units = 16, activation = 'relu', input_dim = 4))

# para a segunda camada oculta não é preciso dizer quantas são as camadas de entrada
rede.add(Dense(units = 16, activation = 'relu'))

# camada de saída, cada saída representa uma classe
# problemas de classificação com mais de duas classes é necessário utilizar o softmax
rede.add(Dense(units = 3, activation = 'softmax'))
```

```python
rede.compile(optimizer = 'adam', loss = 'categorical_crossentropy', metrics = ['categorical_accuracy'])
```

```python
rede.fit(previsores_treinamento, classe_treinamento, batch_size=10, epochs=1000)
```

```python
resultado = rede.evaluate(previsores_teste, classe_teste)
resultado
```

```python
previsoes = rede.predict(previsores_teste)
previsoes
```

```python
previsoes = (previsoes > 0.5)
previsoes
```

### Um artifício muito importante é a matriz de confusão, ela nos diz quantos acertos e erros obtive-mos e em quais classes

### aqui ainda dará um erro pois a variáveis previsoes e classe_teste esta em forma de vetor por exemplo: [0 0 1], e para que seja criada uma matriz  de confusão com esses dados é necessário termos uma variável que mostre exatamente o valor que nos interessa, (o índice de maior valor)

```python
matriz = confusion_matrix(previsoes, classe_teste)
matriz
```

```python
# pegando o maior índice
classe_teste2 = [np.argmax(t) for t in classe_teste]
previsoes2 = [np.argmax(t) for t in previsoes]
print(classe_teste2[:5])
print(previsoes2[:5])
```

```python
# e agora temos o resultado da matriz de confusão, note que em 3x2 temos 3 erros, esse erro só da quando colocamos
# 4 neuronios nas camadas ocultas, mas agora esta com 16 e por isso deu certo 
matriz = confusion_matrix(previsoes2, classe_teste2)
matriz
```
