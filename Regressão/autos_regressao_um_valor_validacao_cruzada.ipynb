### Agora vamos preencher os valores faltantes ###

```python
# criando um dicionário para os valores que quero injetar
valores = {'vehicleType':'limousine','gearbox':'manuell','model':'golf','fuelType':'benzin','notRepairedDamage':'nein'}
```

```python
# injetando os valores nos valores nulos
base = base.fillna(value = valores)
```

## Agora sim vamos para a previsão

```python
previsores = base.iloc[:,1:13].values
```

```python
preco_real = base.iloc[:,0].values
```

### tranformando os valores catégoricos para númericos

```python
from sklearn.preprocessing import LabelEncoder
```

```python
labelencoder_previsores = LabelEncoder()
```

```python
previsores[:,0] = labelencoder_previsores.fit_transform(previsores[:,0])
previsores[:,1] = labelencoder_previsores.fit_transform(previsores[:,1])
previsores[:,3] = labelencoder_previsores.fit_transform(previsores[:,3])
previsores[:,5] = labelencoder_previsores.fit_transform(previsores[:,5])
previsores[:,8] = labelencoder_previsores.fit_transform(previsores[:,8])
previsores[:,9] = labelencoder_previsores.fit_transform(previsores[:,9])
previsores[:,10] = labelencoder_previsores.fit_transform(previsores[:,10])
```

```python
previsores
```

### One hot encoder => é a criação de variáveis do tipo Dummy, o tipo de dados Dummy só devem ser usados em dados categóricos que não tenham prioridade, por exemplo o tipo de câmbio, manual ou automatico, não sabemos qual dos dois é o melhor ###

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
oneHotEncoder = OneHotEncoder(categorical_features= [0,1,3,5,8,9,10])
```

```python
previsores = oneHotEncoder.fit_transform(previsores).toarray()
```

```python
previsores.shape
```

### criando a rede neural ###

```python
from keras.layers import Dense
from keras.models import Sequential
```

```python
def criar_rede():
    regressor = Sequential()
    # 1ª camada oculta, com 316 entradas
    regressor.add(Dense(units = 158, activation= 'relu', input_dim = 316))

    # 2ª camada oculta
    regressor.add(Dense(units = 158, activation= 'relu'))

    # camada de saída, vai gerar somente um número de saída, então só terá uma saída
    # e a função linear não altera o valor de saída
    regressor.add(Dense(units = 1, activation= 'linear'))

    regressor.compile(loss = 'mean_absolute_error', optimizer = 'adam', metrics = ['mean_absolute_error'])
    
    return regressor 
```

## Validação cruzada ##

```python
from sklearn.model_selection import cross_val_score
from keras.wrappers.scikit_learn import KerasRegressor
```

```python
regressor = KerasRegressor(build_fn=criar_rede, 
                          epochs = 100,
                          batch_size = 100)
```

```python
resultados = cross_val_score(estimator=regressor,
                            X = previsores,
                            y = preco_real,
                            cv = 10,
                            scoring = 'mean_absolute_error')
```

```python
media = resultados.mean()
desvio = resultados.std()
```
